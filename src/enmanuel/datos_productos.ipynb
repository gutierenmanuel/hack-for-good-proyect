```python
import gzip
import shutil

# Ruta del archivo gz
archivo_gz = '/home/tr4shhh/Desktop/Proyects/Telefonica_Hackatonn/hack-for-good-proyect/data/raw/en.openfoodfacts.org.products.csv.gz'

# Ruta donde se guardará el archivo descomprimido
archivo_descomprimido = '/home/tr4shhh/Desktop/Proyects/Telefonica_Hackatonn/hack-for-good-proyect/data/raw/en.openfoodfacts.org.products.csv'

with gzip.open(archivo_gz, 'rb') as f_in:
    with open(archivo_descomprimido, 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)

print("Archivo descomprimido exitosamente.")
```

```python
ruta_archivo_original = '/home/tr4shhh/Desktop/Proyects/Telefonica_Hackatonn/hack-for-good-proyect/data/raw/en.openfoodfacts.org.products.csv'

ruta_archivo_nuevo = '/home/tr4shhh/Desktop/Proyects/Telefonica_Hackatonn/hack-for-good-proyect/data/raw/1000_lines_en.openfoodfacts.org.products.csv'

num_lineas = 1000

# Leer las primeras 1000 líneas del archivo original
with open(ruta_archivo_original, 'r') as archivo_original:
    lineas = [archivo_original.readline() for _ in range(num_lineas)]

# Escribir las líneas en el nuevo archivo CSV
with open(ruta_archivo_nuevo, 'w') as archivo_nuevo:
    for linea in lineas:
        archivo_nuevo.write(linea)
```
