```python
## Codigo Scraping 04-02-24 
'''
Siempre añadir user_data_dir y browser_executable_path
Añadidas opciones para que no pida contraseña
Cargar extensiones en el navegador como Ublock y VPN

'''
from selenium.webdriver.common.by import By
import undetected_chromedriver as uc

options = uc.ChromeOptions()
options.headless = False  # Cambia a True si deseas que sea headless (sin interfaz gráfica)
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--password-store=basic")

options.add_experimental_option(
    "prefs",
    {
        "credentials_enable_service": False,
        "profile.password_manager_enabled": False,
    }
)

driver = uc.Chrome(options = options,
                user_data_dir='/home/tr4shhh/Desktop/Proyects/Chrome_profile/Perfil_uc_portatil/', # CUALQUIER CARPETA PERFIL
                browser_executable_path='/usr/bin/chromium-browser', # RUTA DEL EJECUTABLE DEL NAVEGADOR
                )

# Establecer el tamaño de la pantalla (ancho x alto)
driver.set_window_size(1200, 800)
```

```python
driver.get('https://hackforgood.net/category/retos/')
```

```python
xpath_articulo = '/html/body/div[2]/div[1]/div/div/div/main/div/article'


retos = driver.find_elements(By.XPATH, xpath_articulo)
```

```python
xpath_enlace = 'header/h2/a'  # Cambiado a XPath relativo para encontrar el enlace dentro de cada "reto"

enlaces_retos = []

for reto in retos:
    enlace = reto.find_element(By.XPATH, xpath_enlace)
    url_enlace = enlace.get_attribute('href')  # Obtener el atributo href del enlace
    
    enlaces_retos.append(url_enlace)
    print(url_enlace)
```

```python
driver.get(enlaces_retos[0])
```

```python

import time

titulos_retos = []
descripciones = []

for enlace in enlaces_retos:
    driver.get(enlace)

    time.sleep(0.2)

    titulo = driver.find_element(By.XPATH, '/html/body/div[2]/div[1]/div/div/div[1]/main/article/div/header/h2')
    descripcion = driver.find_element(By.XPATH, '/html/body/div[2]/div[1]/div/div/div[1]/main/article/div/div')


    titulos_retos.append(titulo.text)
    descripciones.append(descripcion.text)
```

```python
import pandas as pd

df = pd.DataFrame({'titulo': titulos_retos,
                   'descripcion': descripciones,
                   'enlace': enlaces_retos})
```
